## Proposition

When $ a,b,c \in \mathbb{Z}, a | b, b | c \implies a | c$.

## Proof
Having
$
a \mid b \implies b = ax_1,
b \mid c \implies c = bx_2
$.

Expressing $c = ax_1x_2$ we get $c = ax_3, x_3 = x_1x_2$.

Since the product of two integers is always an integer, then $x_3 \in \mathbb{Z}$.

Therefore: $a \mid c, c = ax_3$  $\blacksquare$

## Proposition
$ n \in N, 2^0 + 2^1 + 2^2 + ... + 2^n = 2^{n+1} - 1$

## Proof
$2^n = sum_{i=0}^{n-1} 2^i + 1$

$2^{n+1}= sum_{i=0}^{n} 2^i + 1$

add `-1` to both sides

$2^{n+1} - 1 =  sum_{i=0}^{n} 2^i + 1 - 1$

$sum_{i=0}^{n} 2^i = 2^{n+1} - 1$

$\blacksquare$
